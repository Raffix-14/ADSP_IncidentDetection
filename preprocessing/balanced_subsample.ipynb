```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_pickle("../data/multi_label_train_v0.3.pkl")
df
```

```python
df = df[df.downloadable == True]
df
```

```python
incidents = df["incidents_list"].value_counts()[1:]
places = df["places_list"].value_counts()[1:]

relative_incident_frequency = incidents / incidents.sum()
relative_place_frequency = places / places.sum()

n = 1000

# Change the random seed of numpy to n
np.random.seed(n)

#Â Get n samples from df maintaining the relative frequency of incidents
incident_samples = np.random.choice(relative_incident_frequency.index, size=n // 2, p=relative_incident_frequency.values)
incident_samples.sort()

places_samples = np.random.choice(relative_place_frequency.index, size=n // 2, p=relative_place_frequency.values)
places_samples.sort()

incident_samples
```

```python
incident_sample_counts = pd.Series(incident_samples).value_counts()
places_sample_counts = pd.Series(places_samples).value_counts()

grouped_incidents = df.groupby("incidents_list")
grouped_places = df.groupby("places_list")

sample_df = pd.DataFrame()

for i in incident_sample_counts.index:
    amount = incident_sample_counts[i]
    sub_df = grouped_incidents.get_group(i).sample(incident_sample_counts[i]) if len(grouped_incidents.get_group(i)) > amount else grouped_incidents.get_group(i)
    sample_df = pd.concat([sample_df, sub_df])

for i in places_sample_counts.index:
    amount = places_sample_counts[i]
    sub_df = grouped_places.get_group(i).sample(places_sample_counts[i]) if len(grouped_places.get_group(i)) > amount else grouped_places.get_group(i)
    sample_df = pd.concat([sample_df, sub_df])

sample_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distrubution of incidents and places
incidents = sample_df["incidents_list"].str.split(", ").explode().value_counts()
places = sample_df["places_list"].str.split(", ").explode().value_counts()

fig, ax = plt.subplots(1, 2, figsize=(20, 20))
sns.barplot(x=incidents.values, y=incidents.index, ax=ax[0])
sns.barplot(x=places.values, y=places.index, ax=ax[1])
ax[0].set_title("Incidents distribution")
ax[1].set_title("Places distribution")
plt.show()

```

```python
alternative_sample_df = df[(df["incidents_list"] != "unknown") & (df["places_list"] != "unknown")].sample(n).copy()

incidents = alternative_sample_df["incidents_list"].str.split(", ").explode().value_counts()
places = alternative_sample_df["places_list"].str.split(", ").explode().value_counts()

fig, ax = plt.subplots(1, 2, figsize=(20, 20))
sns.barplot(x=incidents.values, y=incidents.index, ax=ax[0])
sns.barplot(x=places.values, y=places.index, ax=ax[1])
ax[0].set_title("Incidents distribution")
ax[1].set_title("Places distribution")
plt.show()
```
